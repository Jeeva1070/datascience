# Line chart

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 

a = np.linspace(0,10,20)
b = np.linspace(10,0,20)

fig = plt.figure(figsize = (10,5))

ax1 = plt.subplot()

ax1.plot(a)
ax1.plot(b)

plt.show()
```

# Bar Plot

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 

data = {'apples':10,'oranges':15,'lemons': 10, 'limes':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (5,10))

ax1= plt.subplot()

ax1.bar(names,values)

plt.xlabel('Names')
plt.ylabel('Values')
plt.title('BAR PLOT ')

plt.show()
```

# Scatter Ploting

# Box plot

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 

total = [20,4,1,30,20,10,20,70,30,10]
orders = [10,3,1,15,17,2,30,44,2,1]
discount =[30,20,10,5,20,10,60,20,40,1]

data = list([total,orders,discount])

fig = plt.figure(figsize = (5,5))

ax1 = plt.subplot()

ax1.boxplot(data)


plt.show()


```

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 

total = [20,4,1,30,20,10,20,70,30,10]
orders = [10,3,1,15,17,2,30,44,2,1]
discount =[30,20,10,5,20,10,60,20,40,1]
data = list([total,orders,discount])
fig = plt.figure(figsize = (7,7))
ax1 = plt.subplot()
ax1.boxplot(data)
plt.title('Box Plot')
plt.grid(True)

plt.show()
```
