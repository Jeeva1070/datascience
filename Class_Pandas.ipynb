# Merge Join and Concatenate using DataFrame 

Merge and Join combines the given data into new columns 
Concatenation Combines data in multiple DataFrames without any gaps

```python
leftdf = pd.DataFrame({'Alphabet':['a','b','c','d'],
                       'Number':['1','2','3','4'],
                       'alphanumeric':['a1','b2','c3','d4']})

print(leftdf)
print()
rightdf =pd.DataFrame({'Alphabet':['a','b','c','d'],
                       'Number':['5','6','7','8'],
                       'alphanumeric':['e5','f6','g7','h8']})

print(rightdf)
print()

mergedf = pd.merge(leftdf,rightdf, on ='Alphabet')
mergedf
```

# Join

```python
df1 = pd.DataFrame({'keys': ['0','1','2','3'],
                   'values':['A0','A1','A2','A4']})
df2 = pd.DataFrame({'keys': ['5','7'],
                   'values':['A0','A1']})

jo = df1.join(df2,lsuffix='_first_df', rsuffix = '_second_df')
jo
```

# Concatenate

```python
first_df = pd.DataFrame({'name': ['Jane','John', 'Jacob'],
                        'age': [10,20,30],
                        }, index=[0,1,2])

second_df = pd.DataFrame({'name': ['Neha','Peter', 'Monica'],
                        'age': [40,50,60],
                        }, index=[3,4,5])

third_df= pd.DataFrame({'name': ['Jean','Jenny', 'Jessica'],
                        'age': [70,80,90],
                        }, index=[6,7,8])


joined_df = pd.concat([first_df,second_df,third_df])
joined_df
```
